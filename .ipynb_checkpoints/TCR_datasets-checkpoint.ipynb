#### Distribution of generation probabilities

```python
trb = list(set(vdj['CDR3'][vdj['Gene']=='TRB']))
tra = list(set(vdj['CDR3'][vdj['Gene']=='TRA']))

olga_beta  = OlgaModel(chain_folder = "human_T_beta", recomb_type="VDJ")
olga_alpha  = OlgaModel(chain_folder = "human_T_alpha", recomb_type="VJ")
trb_pgen = parmap.map(olga_beta.compute_aa_cdr3_pgen, trb)
tra_pgen = parmap.map(olga_alpha.compute_aa_cdr3_pgen, tra)

trb_result = pd.DataFrame()
trb_result['CDR3'] = trb
trb_result['Pgen'] = trb_pgen
trb_result['-log10 Pgen'] = -np.log10(trb_pgen)

tra_result = pd.DataFrame()
tra_result['CDR3'] = tra
tra_result['Pgen'] = tra_pgen
tra_result['-log10 Pgen'] = -np.log10(tra_pgen)

ax = sn.kdeplot(tra_result["-log10 Pgen"])
ax = sn.kdeplot(trb_result["-log10 Pgen"]).set(title = 'VDJdb: Generation probabilities distribution')
plt.legend(labels=["TRA","TRB"])
```

```python
from Bio import pairwise2
from Bio.Align import substitution_matrices
from Bio.Seq import Seq

def blosum_metrics(sequence_A, sequence_B, **kwargs):
    """
    Performs a global pairwise alignment between two sequences
    using the BLOSUM62 matrix and the Needleman-Wunsch algorithm
    as implemented in Biopython. Returns the alignment and the score.
    """

    #
    matrix_name = kwargs.get('matrix', 'BLOSUM62')
    matrix = substitution_matrices.load(matrix_name)

    gap_open = kwargs.get('gap_open', -10.0)
    gap_extend = kwargs.get('gap_extend', -0.5)

    alns = pairwise2.align.globalds(sequence_A, sequence_B,
                                    matrix, gap_open, gap_extend,
                                    penalize_end_gaps=(False, False))

    best_aln = alns[0]
    aligned_A, aligned_B, score, begin, end = best_aln
    return score
```

```python
vdj = pd.read_csv('SearchTable-2023-05-19 17_34_21.888.tsv', sep = "\t")
trb = list(set(vdj['CDR3'][vdj['Gene']=='TRB']))
tra = list(set(vdj['CDR3'][vdj['Gene']=='TRA']))

cross = vdj[['CDR3', 'Epitope']].drop_duplicates()
cross = pd.DataFrame(cross['CDR3'].value_counts())
cross = cross.reset_index()
cross.columns = ['CDR3', '# of epitopes']

olga_beta  = OlgaModel(chain_folder = "human_T_beta", recomb_type="VDJ")
olga_alpha  = OlgaModel(chain_folder = "human_T_alpha", recomb_type="VJ")
trb_pgen = parmap.map(olga_beta.compute_aa_cdr3_pgen, trb)
tra_pgen = parmap.map(olga_alpha.compute_aa_cdr3_pgen, tra)

trb_result = pd.DataFrame()
trb_result['CDR3'] = trb
trb_result['Pgen'] = trb_pgen
trb_result['-log10 Pgen'] = -np.log10(trb_pgen)

tra_result = pd.DataFrame()
tra_result['CDR3'] = tra
tra_result['Pgen'] = tra_pgen
tra_result['-log10 Pgen'] = -np.log10(tra_pgen)

cross_pgen_trb = trb_result.merge(cross, on = 'CDR3')
cross_pgen_tra = tra_result.merge(cross, on = 'CDR3')
```

```python
ax = sn.relplot(data=cross_pgen_trb, x="-log10 Pgen", y="# of epitopes")
ax.fig.suptitle('TRB')
```

```python
mpas = pd.read_csv('McPAS-TCR_human.csv')
mpas = mpas[['CDR3.alpha.aa', 'CDR3.beta.aa', 'Epitope.peptide']]
mpas = mpas.dropna()

olga_beta  = OlgaModel(chain_folder = "human_T_beta", recomb_type="VDJ")
trb_mpas = set(mpas['CDR3.beta.aa'])
olga_alpha  = OlgaModel(chain_folder = "human_T_alpha", recomb_type="VJ")
tra_mpas = set(mpas['CDR3.alpha.aa'])

trb_mpas_pgen = parmap.map(olga_beta.compute_aa_cdr3_pgen, trb_mpas)
tra_mpas_pgen = parmap.map(olga_alpha.compute_aa_cdr3_pgen, tra_mpas)

trb_mpas_result = pd.DataFrame()
trb_mpas_result['CDR3.beta.aa'] = list(trb_mpas)
trb_mpas_result['Pgen TRB'] = trb_mpas_pgen
trb_mpas_result['-log10 Pgen TRB'] = -np.log10(trb_mpas_pgen)

tra_mpas_result = pd.DataFrame()
tra_mpas_result['CDR3.alpha.aa'] = list(tra_mpas)
tra_mpas_result['Pgen TRA'] = tra_mpas_pgen
tra_mpas_result['-log10 Pgen TRA'] = -np.log10(tra_mpas_pgen)

mpas = mpas.merge(tra_mpas_result, on = 'CDR3.alpha.aa')
mpas = mpas.merge(trb_mpas_result, on = 'CDR3.beta.aa')
mpas['Pgen'] = mpas['Pgen TRA'] * mpas['Pgen TRB']
mpas['-log10 Pgen'] = -np.log10(mpas['Pgen'])
mpas['CDR3.aa'] = mpas['CDR3.alpha.aa'] + mpas['CDR3.beta.aa']

cross_mpas = mpas[['CDR3.aa', 'Epitope.peptide']].drop_duplicates()
cross_mpas = pd.DataFrame(cross_mpas['CDR3.aa'].value_counts())
cross_mpas = cross_mpas.reset_index()
cross_mpas.columns = ['CDR3.aa', '# of epitopes']
mpas = mpas.merge(cross_mpas, on = 'CDR3.aa')

cross_mpas_tra = mpas[['CDR3.alpha.aa', 'Epitope.peptide']].drop_duplicates()
cross_mpas_tra = pd.DataFrame(cross_mpas_tra['CDR3.alpha.aa'].value_counts())
cross_mpas_tra = cross_mpas_tra.reset_index()
cross_mpas_tra.columns = ['CDR3.alpha.aa', '# of epitopes TRA']
mpas = mpas.merge(cross_mpas_tra, on = 'CDR3.alpha.aa')

cross_mpas_trb = mpas[['CDR3.beta.aa', 'Epitope.peptide']].drop_duplicates()
cross_mpas_trb = pd.DataFrame(cross_mpas_trb['CDR3.beta.aa'].value_counts())
cross_mpas_trb = cross_mpas_trb.reset_index()
cross_mpas_trb.columns = ['CDR3.beta.aa', '# of epitopes TRB']
mpas = mpas.merge(cross_mpas_trb, on = 'CDR3.beta.aa')
```
